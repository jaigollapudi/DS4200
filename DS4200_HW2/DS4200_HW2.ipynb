### Part 3 - Basic data analysis with Python

Here is a dataset of Seattle weather in 4 years. Load the data and generate corresponding data visualizations. Make sure each figure has title, labels and legends if necessary. Add one or two sentences in each figure to talk about what you have learned from the figure. 

```python
from vega_datasets import data

weather = data.seattle_weather()
weather.head()
```

```python
# Importing Libraries
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import altair as alt
```

##### Question 1 (5 pts)

Generate a data visualization to show the relationship between precipitation and max temperature.  

```python
# Generating scatterplot and setting title and labels
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temp_max', y='precipitation', data=weather)
plt.title('Precipitation vs. Max Temperature in Seattle', fontsize=12)
plt.xlabel('Max Temperature (C)', fontsize=12)
plt.ylabel('Precipitation (mm)', fontsize=12)
plt.show()
```

**What I learned:**

This scatter plot shows how precipitation in Seattle relates to the maximum temperature. Most of the data points are clustered at the lower end of the precipitation scale, indicating that on most days, the precipitation is minimal regardless of the temperature. However, as the temperature increases, the frequency of days with higher precipitation decreases, which is especially noticeable above 20C. This could suggest that warmer days in Seattle are typically drier. 

##### Question 2 (5 pts)

Generate a data visualization to show the relationship between max temperature and weather type.  

```python
# Generating boxplot and setting title and labels
plt.figure(figsize=(10, 6))
sns.boxplot(x='weather', y='temp_max', data=weather)
plt.title('Max Temperature Distribution by Weather Type in Seattle', fontsize=12)
plt.xlabel('Weather Type', fontsize=12)
plt.ylabel('Max Temperature (C)', fontsize=12)

# Rotating x-axis by 45 degrees
plt.xticks(rotation=45)
plt.show()
```

**What I learned:**

The visualization is a box plot displaying the distribution of maximum temperatures for different weather types in Seattle. From the chart, we can observe the presence of outliers, especially in the 'Rain' and 'Fog' categories, suggesting that there are days with extreme temperatures that fall outside the typical range for those weather types. The following are a few more observations:

- Snow: Snowy days have the lowest median max temperature, and the range is quite narrow, which means that snowy days in Seattle are consistently cold.
- Sun: Sunny days have the highest median max temperature and the second largest spread of data, including several warm outliers, indicating that the warmest days in Seattle tend to be sunny.

##### Question 3 (5 pts)

Generate a data visualization to show the relationship between max temperature and min temperature, and then show different color for differnet weather type. 

```python
# Generating scatterplot and setting title, labels, and legend
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temp_min', y='temp_max', hue='weather', data=weather)
plt.title('Max vs. Min Temperature by Weather Type in Seattle', fontsize=12)
plt.xlabel('Min Temperature (C)', fontsize=12)
plt.ylabel('Max Temperature (C)', fontsize=12)
plt.legend(title='Weather Type')
plt.show()
```

**What I learned:**

The scatter plot visualizes the relationship between the maximum and minimum temperatures in Seattle, with data points colored according to the weather type. From the plot, we can observe several trends:

- Positive Correlation: There is a clear positive correlation between the maximum and minimum temperatures. As the minimum temperature increases, the maximum temperature tends to increase as well.

- Weather Types: Different weather types are represented by different colors. It seems that sunny days (in green) tend to have a wider range of temperatures, possibly indicating that sunny weather is associated with both mild and hot days. Days with snow (in red) are clustered at the lower end of both temperature scales, indicating that snowy days are consistently cold.

- Dense Clusters: There are dense clusters of days with similar minimum and maximum temperatures for each weather type, especially noticeable with fog and snow.

This visualization provides a detailed look at how different weather types distribute across temperature ranges in Seattle, which can be useful for understanding weather patterns and planning activities based on typical weather conditions.

##### Question 4 (5 pts) 

Generate a data visualization to show the relationship between max temperature, min temperature and wind speed. 

```python
# Adding subplot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Generating scatter plot
sc = ax.scatter(weather['temp_max'], weather['temp_min'], weather['wind'], c=weather['wind'], cmap='viridis', marker='o')

# Adding a color bar
cbar = plt.colorbar(sc, ax=ax, shrink=0.5, aspect=5, pad=0.2)  
cbar.set_label('Wind Speed (m/s)', fontsize=12)

# Setting title and labels
ax.set_title('3D Plot of Max Temperature, Min Temperature, and Wind Speed in Seattle', fontsize=14)
ax.set_xlabel('Max Temperature (C)', fontsize=12)
ax.set_ylabel('Min Temperature (C)', fontsize=12)
ax.set_zlabel('Wind Speed (m/s)', fontsize=12)

plt.show()
```

**What I learned:**

This visualization depicts a 3D scatter plot that represents the relationship between maximum temperature, minimum temperature, and wind speed in Seattle. The following are some observations:

- Temperature Relationship: There is a clear positive relationship between max and min temperatures; as the min temperature increases, the max temperature tends to increase as well, forming a cloud of points along a diagonal plane in the 3D space.
- Wind Speed Variation: Wind speed is represented in the vertical dimension, and there appears to be a wide range of wind speeds across different temperatures. 
- Data Distribution: The bulk of data points are concentrated in the area with moderate temperatures and wind speeds, which suggests these are the most common conditions in Seattle.

### Part 4 - Basic data analysis with Altair

We are going to use the same data but use Altair to create the data visualziation. 

##### Question 1 (5 points)

Recreate the figure in Part 3 Question 2 with Altair. 

```python
# Recreating Part 3, Q2 using Altair 
chart = alt.Chart(weather).mark_boxplot().encode(
    x=alt.X('weather:N', title = 'Weather Type', axis=alt.Axis(labelFontSize=12, labelAngle=45)), 
    y=alt.Y('temp_max:Q', title= 'Max Temperature (C)', axis=alt.Axis(labelFontSize=12)), 
    color=alt.Color('weather:N', title='Weather Type', scale=alt.Scale(scheme= 'set1')),
    tooltip=['weather', 'temp_max']
).properties(
    title= alt.TitleParams('Max Temperature Distribution by Weather Type in Seattle', fontSize=16),
    width=800,
    height=400
)

chart.display()
```

**What I learned:**

The visualization shows the distribution of maximum temperatures by different weather types in Seattle. It represents five categories of weather: drizzle, fog, rain, snow, and sun. Overall, the box plot effectively communicates the variability and central tendencies of maximum temperatures associated with different weather types, allowing viewers to easily compare weather conditions in Seattle using the interactive feature.

##### Question 2 (5 points)

Recreate the figure in Part 3 Question 3 with Altair.

```python
# Recreating Part 3, Q3 using Altair
scatter_chart = alt.Chart(weather).mark_circle(size=60).encode(
    x=alt.X('temp_min:Q', axis=alt.Axis(title='Min Temperature (C)', titleFontSize=14, labelFontSize=12)),
    y=alt.Y('temp_max:Q', axis=alt.Axis(title='Max Temperature (C)', titleFontSize=14, labelFontSize=12)),
    color=alt.Color('weather:N', legend=alt.Legend(title='Weather Type', titleFontSize=12, labelFontSize=10)),
    tooltip=['temp_min', 'temp_max', 'weather']
).properties(
    title=alt.TitleParams('Max vs. Min Temperature by Weather Type in Seattle', fontSize=16),
    width=800,
    height=400
)

scatter_chart.display()
```

**What I learned:**

This scatter plot visualizes the relationship between maximum and minimum temperatures, colored by weather type. It allows for easy identification of patterns, such as which weather types are common at different temperature ranges, enhancing understanding of Seattle's climate dynamics using the interactive features

##### Question 3 (10 points)

Create a facet figure with 2 row and 3 columns. For each panel in the facet figure, it shows a scatter plot between the max and min temperature for one type of weather. Use the wind speed to color the points. Also add interaction to show the max and min temperature and wind speed for each data point. Include necessary title and labels. 

```python
# Creating facet chart
facet_chart = alt.Chart(weather).mark_circle().encode(
    x=alt.X('temp_min:Q', axis=alt.Axis(title='Min Temperature (C)', titleFontSize=14, labelFontSize=12)),
    y=alt.Y('temp_max:Q', axis=alt.Axis(title='Max Temperature (C)', titleFontSize=14, labelFontSize=12)),
    color=alt.Color('wind:Q', legend=alt.Legend(title='Wind Speed (m/s)', titleFontSize=12, labelFontSize=10)),
    tooltip=['temp_min', 'temp_max', 'wind']
).properties(
    width=200,
    height=200
).facet(
    facet=alt.Facet('weather:N', title='Weather Type', header=alt.Header(titleFontSize=14, labelFontSize=12)),
    columns=3
).properties(
    title=alt.TitleParams('Scatter Plot of Max vs. Min Temperature for Each Weather Type in Seattle', fontSize=16)
)

facet_chart.display()
```

This faceted chart creates individual scatter plots for each weather type, showing the relationship between max and min temperatures, with points colored by wind speed. This visualization enables detailed comparison across weather types and how wind speed correlates with temperature variations within each weather category.



##### Question 4 (10 points)

Create a compound figure with at least two types of marks. Show the trend of average temperature (the average of min and max temperature) in 4 years. Include necessary title and labels.

```python
# Calculating the average temperature
weather['avg_temp'] = (weather['temp_max'] + weather['temp_min']) / 2

# Creating the visualization
line_chart = alt.Chart(weather).mark_line(color='red').encode(
    x='year(date):T',
    y='mean(avg_temp):Q',
    tooltip=['year(date)', 'mean(avg_temp)']
)

points = alt.Chart(weather).mark_point(color='blue').encode(
    x='year(date):T',
    y='mean(avg_temp):Q',
    size=alt.value(100),
    tooltip=['year(date)', 'mean(avg_temp)']
)

compound_chart = (line_chart + points).properties(
    title=alt.TitleParams('Trend of Average Temperature in Seattle Over 4 Years', fontSize=16)
)

compound_chart.display()
```
